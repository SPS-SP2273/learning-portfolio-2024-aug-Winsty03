```python
%%timeit
squares = [i**2 for i in range(10)]

```

It takes 840 nanoseconds for the command to execute once. 96.6 nanoseconds is the standard deviationof the distribution, showing its variation. Each loop is run 7 times and there was 1,000,000 loops in this case. 

```python
print(squares)
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
    
```

It takes 2.14 microseconds for the command to execute once. Hence, the first command runs faster than the second command. The second number of 133 nanoseconds is the standard deviation of the distribution, hence showing its variation. Each loop is run 7 times and there was 100,000 loops in this case.


```python
print(squares)
```
